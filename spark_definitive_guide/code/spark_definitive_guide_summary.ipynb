```python
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName('Spark-Definitive-Guide').getOrCreate()

BASEPATH = "/Users/zhuohuawu/Documents/data/spark-definitive-guide"
df = spark.read.format("json").load(BASEPATH + "/data/flight-data/json/2015-summary.json")
df.show(5, False)
```

```python
from pyspark.sql.functions import desc
df.groupBy("DEST_COUNTRY_NAME", "ORIGIN_COUNTRY_NAME").count().orderBy(desc("count")).show(200, False)
```
